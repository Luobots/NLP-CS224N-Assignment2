```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []
with open('..\训练细节.txt') as file:
    for word in file:
        word = word.replace(':', '').split()
        x.append(int(word[1]))
        y.append(float(word[2]))
```

```python
plt.ylabel('loss')
plt.xlabel('iteration')
plt.plot(x, y)
plt.savefig('overall loss map.png')
plt.show()
```

```python
plt.ylabel('loss')
plt.xlabel('iteration')
plt.plot(x[2000: 4000], y[2000: 4000])
plt.savefig('20000-40000 loss map.png')
plt.show()
```
